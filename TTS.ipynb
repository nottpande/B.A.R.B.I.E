```python
from transformers import VitsModel, AutoTokenizer
import torch

model = VitsModel.from_pretrained("facebook/mms-tts-kan")
tokenizer = AutoTokenizer.from_pretrained("facebook/mms-tts-kan")

text = "ಕನ್ನಡ ವಿಕಿಪೀಡಿಯ ಕನ್ನಡದ ಒಂದು ಸ್ವತಂತ್ರ ವಿಶ್ವಕೋಶ."
inputs = tokenizer(text, return_tensors="pt")

with torch.no_grad():
    output = model(**inputs).waveform
```

```python
import scipy.io.wavfile as wav
import numpy as np

# Move output to CPU if on GPU and convert to numpy
output_np = output.cpu().numpy()

# Ensure the output is one-dimensional (mono) or two-dimensional (stereo)
# Reshape if necessary (assuming mono for simplicity)
if output_np.ndim > 1:
    output_np = output_np.squeeze()

# Scale to 16-bit PCM range and cast to int16
output_scaled = np.int16(output_np * 32767)

# Set a slower playback sampling rate by reducing from 44100 Hz (e.g., try 22050 Hz or 16000 Hz)
slower_sampling_rate = 22050  # Adjust this value for desired speed

# Write to WAV file with the slower sampling rate
wav.write("techno.wav", rate=slower_sampling_rate, data=output_scaled)
```
